### 2.3 Fixing outliers

```python
data.describe()
```

#### What we can do:
1) ignore outliers (keeping all rows)
2) removing outilers (deleting rows where we have those extreme values)
3) handling outilers (capping the outilers to some normal value in the data)
   
KEY POINTS
1) Removing outliers is suitable when they are clearly errors or irrelevant to the analysis, while capping is preferred when you want to mitigate the influence of extreme values without losing data
2) If your goal is robust business reporting (like calculating , trend or time series analysis), capping is often better than deletion, as long as it’s documented and well-thought-out.
3) If your goal is anomaly detection, fraud analysis, or machine learning, avoid capping, and instead flag outliers for analysis.

so we assuming that there r outliers and we do either 2 or 3 for fixing them based on outlier we have in each column(numeric col)

```python
# in col is_canceled we dont have outliers as we can see that it has only 2 values(0 or 1) in which 25% we have 0 and 75% we have 1 which is fine
```

```python
# in col lead_time we have outlier ,see max value which is 737 days of advance booking (nearly 2 years) that looks suspicious (though VERY few hotels allow 2 years of advance booking but coz of that VERY few it can consider as outlier)
# we deleting outliers for this col coz if we do capping then it may affect our result 
data[data['lead_time']>547].count().sum() 
# so we can consider max advance booking upto 18 months(547 days) (in general it is 1 year(12 month) but here we have more than 1 lakh rows that booked 12 month in advance so we cant delete that many rows as we loose lot of informative data)
# for more than 18 months advance booking we have 9044 rows which we can delete (we can also cap the data but here we removing coz we assuming that more than 18 month advance booking is error or irrelevant)
print('before deleting')
plt.boxplot(data['lead_time'])
plt.show()

data=data[data['lead_time']<=547]
print('after deleting')
plt.boxplot(data['lead_time'])
plt.show()
```

```python
data.describe()
```

```python
#lead_time has relation with arrival date year,week number,and day od month columns (ie booking to some date means it is arrival date) so no changes in those col
```

```python
#for stays_in_weekend_nights and stays_in_week_nights col
# max value for weekend nights is 19 and for week nights is 50 which are outlier (as thats lot of days to stay in hotel which is VERY rare)
print('before fixing outlier')
plt.boxplot([data['stays_in_weekend_nights'],data['stays_in_week_nights']])
plt.show()

#we assuming max can stay in weekend is 10 and week is 30 as it is not that rare case as before
#we will cap the outliers (ie in weekend if value is more than 10 then cap to 10 and and in week if value is more than 30 then cap to 30)
#in report mention if the result is 30 in week or 10 in weekend then it represent those how stayed >=30 or >=10 not just 30 or 10
data['stays_in_weekend_nights']=np.where(data['stays_in_weekend_nights']>10,10,data['stays_in_weekend_nights'])
data['stays_in_week_nights']=np.where(data['stays_in_week_nights']>30,30,data['stays_in_week_nights'])
print('after fixing outlier')
plt.boxplot([data['stays_in_weekend_nights'],data['stays_in_week_nights']])
plt.show()
```

## 3. Feature engineering

Feature engineering is the process of selecting, transforming, and creating new features (columns) from raw data or existing features (without deleting those existing features).

It helps to improve the performance of machine learning models

```python
data.info()
```

```python
# from col arrival_date_year, arrival_date_month and arrival_date_day_of_month we creating new col as arrival_date
index=data.columns.get_loc('lead_time')
arrival_date=data['arrival_date_year'].astype('str')+'-'+data['arrival_date_month']+'-'+data['arrival_date_day_of_month'].astype('str')
data.insert(index+1,'arrival_date',pd.to_datetime(arrival_date,format='%Y-%B-%d')) #inserting after col lead_time
data.head(5)
```

```python
#from stays_in_weekend_nights and stays in week_nights we create new column as total_stay_nights
index=data.columns.get_loc('arrival_date_day_of_month')
total_stay_nights=data['stays_in_weekend_nights']+data['stays_in_week_nights']
data.insert(index+1,'total_stay_nights',total_stay_nights)
data.head(5)
```

```python
#from adults, children and babies creating new column as total_guest
index=data.columns.get_loc('stays_in_week_nights')
total_guests=data['adults']+data['children']+data['babies']
data.insert(index+1,'total_guests',total_guests)
data.head(5)
```

```python
# from total_stay_nights and adr we creating new col as revenue (revenue=tota stay*adr)
index=data.columns.get_loc('adr')
revenue=data['total_stay_nights']*data['adr']
data.insert(index+1,'revenue',revenue)
data.head(5)
```
